```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-fr-en")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-fr-en")

input_text = "Ce cours est produit par Hugging Face."
tokenized_input = tokenizer(input_text, return_tensors="pt")

output_tokens = model.generate(**tokenized_input)
translated_text = tokenizer.decode(output_tokens[0], skip_special_tokens=True)
print(translated_text)
```

```python
import librosa
import requests
import numpy as np
from transformers import AutoProcessor, AutoModelForSpeechSeq2Seq

# Download the audio file
audio_url = "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac"
local_audio_path = "/tmp/mlk.flac"

response = requests.get(audio_url)
response.raise_for_status()  # Raise an exception for HTTP errors
with open(local_audio_path, "wb") as f:
    f.write(response.content)

# Load the audio file using librosa, resample to 16kHz
speech_array, sampling_rate = librosa.load(local_audio_path, sr=16000)

# Load the processor and model
processor = AutoProcessor.from_pretrained("openai/whisper-base.en")
model = AutoModelForSpeechSeq2Seq.from_pretrained("openai/whisper-base.en")

# Prepare the audio input for the model
input_features = processor(speech_array, sampling_rate=sampling_rate, return_tensors="pt").input_features

# Generate output token ids
generated_ids = model.generate(input_features)

# Decode the transcription
transcription = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(transcription)
```

```python
import librosa
import requests
import numpy as np
from transformers import AutoProcessor, AutoModelForSpeechSeq2Seq

# Download the audio file
audio_url = "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac"
local_audio_path = "/tmp/mlk.flac"

response = requests.get(audio_url)
response.raise_for_status()  # Raise an exception for HTTP errors
with open(local_audio_path, "wb") as f:
    f.write(response.content)

# Load the audio file using librosa, resample to 16kHz
speech_array, sampling_rate = librosa.load(local_audio_path, sr=16000)

# Load the processor and model
processor = AutoProcessor.from_pretrained("openai/whisper-base.en")
model = AutoModelForSpeechSeq2Seq.from_pretrained("openai/whisper-base.en")

# Prepare the audio input for the model
input_features = processor(speech_array, sampling_rate=sampling_rate, return_tensors="pt").input_features

# Generate output token ids
generated_ids = model.generate(input_features)

# Decode the transcription
transcription = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(transcription)
```
